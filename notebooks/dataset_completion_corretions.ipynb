## This notebook removes 2014-2019 survey answers from the combined .csv file and re-adds 2014 - 2019 individually to make the file cleaner.

### 1. Remove 2014-2019 rows from current df_combined.

```python
import pandas as pd
```

```python
df_combined = pd.read_csv("../../data/datasets/osmi_mental_health_combined_data.csv")
```

```python
df_combined = df_combined[df_combined["SurveyID"] >= 2020].reset_index(drop=True)
df_combined.shape
```

### Let's re-add each year individually

#### 2014

```python
df_2014 = pd.read_csv("../../data/datasets/osmi_mental_health_2014.csv")

df_2014.columns = df_2014.columns.str.strip()
df_combined.columns = df_combined.columns.str.strip()

common_cols = sorted(list(set(df_combined.columns).intersection(df_2014.columns)))

df_2014["SurveyID"] = 2014

if "SurveyID" not in common_cols:
    common_cols = ["SurveyID"] + common_cols

df_combined_common = df_combined[common_cols].copy()
df_2014_common      = df_2014[common_cols].copy()

df_combined = pd.concat(
    [df_combined_common, df_2014_common],
    ignore_index=True
)
```
